```python
import sys
sys.path.append('../')


import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

import ten
```

```python
# Datos del paper de McNeal

# Perylene:
aceptor_perylene = [0, 21, 43, 86, 171, 436, 871]
eff_perylene = [0, 0.1875, 0.3125, 0.68, 0.78, 0.9, 0.95]
r_perylene = 2.29

# Coumarin 6
aceptor_coumarin = [0, 20, 60, 120, 240, 600, 1000]
eff_coumarin = [0, 0.42, 0.58, 0.85, 0.92, 0.97, 0.98]
r_coumarin = 3.05

# TPP
aceptor_TPP = [0, 40, 100, 200, 400, 1000]
eff_TPP = [0, 0.48, 0.82, 0.9, 0.95, 0.98]
r_TPP = 3.14
```

```python
#y1 = eff_perylene
#y2 = eff_coumarin
#y3 = eff_TPP
```

```python
data_per = ten.post.extrac4dir('data/perilene/', aceptor_perylene)
data_per = ten.post.diference2(eff_perylene, data_per)
element_min_per = np.argmin(data_per[:,3])

D_per = data_per[element_min_per][0]
epsilon_per = data_per[element_min_per][1]

data_cou = ten.post.extrac4dir('data/coumarin/', aceptor_coumarin)
data_cou = ten.post.diference2(eff_coumarin, data_cou)
element_min_cou = np.argmin(data_cou[:,3])

D_cou = data_cou[element_min_cou][0]
epsilon_cou = data_cou[element_min_cou][1]

data_TPP = ten.post.extrac4dir('data/TPP/', aceptor_TPP)
data_TPP = ten.post.diference2(eff_TPP, data_TPP)
element_min_TPP = np.argmin(data_TPP[:,3])

D_TPP = data_TPP[element_min_TPP][0]
epsilon_TPP = data_TPP[element_min_TPP][1]
```

```python
x1 = aceptor_perylene
x2 = aceptor_coumarin
x3 = aceptor_TPP

for i in data_per:
    if i[0] == 40 and i[1] == 1.5:
        y_per = i[4:]
for i in data_cou:
    if i[0] == 40 and i[1] == 1.5:
        y_cou = i[4:]
for i in data_TPP:
    if i[0] == 40 and i[1] == 1.5:
        y_tpp = i[4:]
```

```python
fig = plt.figure(figsize=(20, 15), dpi=300)
plt.subplots_adjust(hspace=0.000001)

# PERILENE
ax1 = plt.subplot(3, 1, 1)
ax1.plot(x1, eff_perylene, 'b--o', label='Experimental', linewidth=4, ms=12)
ax1.plot(x1, y_per, 'r-.s', label='Simulate',linewidth=4, ms=12)
#ax1.legend(loc=4, ncol=2)
ax1.text(4, 0.85, r'$Perylene$', fontsize=30)
ax1.text(0, 0.75, r'$R_{foster} = %.3f \; nm$' %(r_perylene), fontsize=20)
#plt.xlim((-10, 210))
plt.ylim((-0.1, 1))
plt.yticks(np.arange(0, 1, 0.1), fontsize=15)
ax1.grid(True)

# COUMARINE
ax2 = plt.subplot(3, 1, 2, sharex=ax1)
ax2.plot(x2, eff_coumarin, 'b--o', label='Experimental', linewidth=4, ms=12)
ax2.plot(x2, y_cou, 'r-.s', label='Simulate', linewidth=4, ms=12)
#ax2.legend(loc=4, ncol=2)
ax2.text(0, 0.85, r'$Coumarine\;6$', fontsize=30)
ax2.text(0, 0.75, r'$R_{foster} = %.3f \; nm$' %(r_coumarin), fontsize=20)
plt.ylabel(r'$Eficiencia $', fontsize=40)
#plt.xlim((-10, 1010))
plt.ylim((-0.1, 1))
plt.yticks(np.arange(0, 1, 0.1), fontsize=15)
ax2.grid(True)

# TPP
ax3 = plt.subplot(3, 1, 3, sharex=ax1)
ax3.plot(x3, eff_TPP, 'b--o', label='Experimental', linewidth=4, ms=12)
ax3.plot(x3, y_tpp, 'r-.s', label='Simulate', linewidth=4, ms=12)
#ax3.legend(loc=4, ncol=2)
ax3.text(10, 0.85, r'$TPP$', fontsize=30)
ax3.text(0, 0.75, r'$R_{foster} = %.3f \; nm$' %(r_TPP), fontsize=20)
plt.xlabel(r'$Número\; de\; aceptores$', fontsize=40)
#plt.xlim((-10, 1010))
plt.ylim((-0.1, 1))
plt.yticks(np.arange(0, 1, 0.1), fontsize=15)
ax3.grid(True)
plt.xlim((-10, 210))
plt.xticks(np.arange(0, 201, 25), fontsize=15)

a = plt.axes([.5, .15, .4, .10])
plt.plot(x1, eff_perylene, 'bo--', x1, y_per, 'r-.s', linewidth=3, ms=8)
plt.xlim((-10, 900))
plt.ylim((-0.1, 1.1))
#plt.xticks([])
#plt.yticks([])
plt.xticks(np.arange(0, 801, 100))

a = plt.axes([.5, .41, .4, .10])
plt.plot(x2, eff_coumarin, 'bo--', x2, y_cou, 'r-.s', linewidth=3, ms=8)
plt.xlim((-10, 1010))
plt.ylim((-0.1, 1.1))
plt.xticks(np.arange(0, 901, 100))

a = plt.axes([.5, .665, .4, .10])
plt.plot(x3, eff_TPP, 'bo--', x3, y_tpp, 'r-.s', linewidth=3, ms=8)
plt.xlim((-10, 1010))
plt.ylim((-0.1, 1.1))
plt.xticks(np.arange(0, 901, 100))


xticklabels = ax1.get_xticklabels() + ax2.get_xticklabels()
plt.setp(xticklabels, visible=False)

plt.savefig('eff*3.svg', transparent=True)
plt.show()
```
